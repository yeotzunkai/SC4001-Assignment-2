```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv('basedata.csv')

# Extract the necessary data
train_losses = df['Training Loss']
valid_losses = df['Validation Loss']
train_accuracies = df['Training Accuracy']  # Adjust this line as needed if your values are not scaled correctly
valid_accuracies = df['Validation Accuracy']  # Adjust this line as needed if your values are not scaled correctly

# Get the last epoch's values for labels
last_epoch_index = len(train_losses) - 1
last_train_loss = train_losses.iloc[-1]
last_valid_loss = valid_losses.iloc[-1]
last_train_accuracy = train_accuracies.iloc[-1]
last_valid_accuracy = valid_accuracies.iloc[-1]

# After training, plot the training and validation losses and accuracies
plt.figure(figsize=(12, 4))

# Plot training and validation loss
plt.subplot(1, 2, 1)
plt.plot(train_losses, label='Training Loss')
plt.plot(valid_losses, label='Validation Loss')
plt.legend()
plt.title('Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
# Display last epoch loss as text
plt.text(last_epoch_index, last_train_loss, f'{last_train_loss:.4f}', ha='right')
plt.text(last_epoch_index, last_valid_loss, f'{last_valid_loss:.4f}', ha='right')

# Plot training and validation accuracy
plt.subplot(1, 2, 2)
plt.plot(train_accuracies, label='Training Accuracy')
plt.plot(valid_accuracies, label='Validation Accuracy')
plt.legend()
plt.title('Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
# Display last epoch accuracy as text
plt.text(last_epoch_index, last_train_accuracy, f'{last_train_accuracy:.2f}%', ha='right')
plt.text(last_epoch_index, last_valid_accuracy, f'{last_valid_accuracy:.2f}%', ha='right')

# Show the plot
plt.show()
```
